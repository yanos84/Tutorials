```python
import numpy as np

x = np.array([1,2,3,7,10])
y  = np.array([3,6.3,12.2,20.4,28.6])
x_mean = np.mean(x)
y_mean = np.mean(y)
print(x_mean)
print(y_mean)
x_std = np.std(x)
y_std = np.std(y)
print(x_std)
print(y_std)
cov_x_y = np.cov(x,y)[0,1]
print(cov_x_y)
print(cov_x_y/(x_std*y_std))
np.cov(x,y)
```

```python
import numpy as np

numpy_correlation_matrix = np.corrcoef(x, y)
numpy_correlation_coefficient = numpy_correlation_matrix[0, 1]

print(f"Pearson correlation coefficient using numpy.corrcoef: {numpy_correlation_coefficient:.4f}")
```

```python
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
import numpy as np

# Get the first three elements of x and y
x_lagrange = x[:3]
y_lagrange = y[:3]

# Compute the Lagrange polynomial
poly = lagrange(x_lagrange, y_lagrange)

print(f"Lagrange polynomial coefficients: {poly.coefficients}")
```

```python
(61.8-20.4)**2
0.6**2
```

To visualize the Lagrange polynomial and the original points:

```python
# Generate a range of x values to plot the polynomial
x_new = np.linspace(x_lagrange.min(), x_lagrange.max(), 100)
y_new = poly(x_new)

plt.figure(figsize=(8, 6))
plt.plot(x_lagrange, y_lagrange, 'o', label='Original Points (first 3)')
plt.plot(x_new, y_new, '-', label='Lagrange Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Interpolation')
plt.legend()
plt.grid(True)
plt.show()
```
