This Python snippet uses a DataFrame to filter and analyze Titanic passenger data by gender. It demonstrates how to extract subgroups and compute descriptive statisticsâ€”skills essential for social science research.


```python
# Filter by gender
women = titanic[titanic['sex'] == 'female']
men = titanic[titanic['sex'] == 'male']

# Compute mean survival rate by sex
print("Female survival rate:", women['survived'].mean())
print("Male survival rate:", men['survived'].mean())
```



---

This snippet demonstrates how to use grouping in a DataFrame to analyze survival rates on the Titanic by social class and genderâ€”a classic intersectional analysis.

groupby(['class', 'sex']) Groups the data by two variables: passenger class (1st, 2nd, 3rd) and gender (male, female).

['survived'].mean() Calculates the average survival rate for each group. Since survived is binary (1 = survived, 0 = did not), the mean gives the proportion who survived.

.reset_index() Converts the grouped data back into a regular DataFrame so it can be displayed or further analyzed.



```python
# Group by class and gender
titanic.groupby(['class', 'sex'])['survived'].mean()
```

This Python snippet uses Seaborn and Matplotlib to create a histogram of passenger ages from the Titanic dataset. Visualizations like this are essential in human science for understanding distributions, spotting patterns, and communicating findings.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=titanic, x="age", kde=True, bins=30, color="skyblue")
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

* sns.histplot(...) Creates a histogram of the age variable, showing how many passengers fall into each age range.

* kde=True Adds a Kernel Density Estimate curveâ€”a smooth line that approximates the distribution. This is useful for identifying peaks and spread.

* bins=30 Divides the age range into 30 intervals (bins), giving a more detailed view of the distribution.

* color="skyblue" Sets the color of the bars for visual appeal and clarity.

* plt.title(...), plt.xlabel(...), plt.ylabel(...) Adds a title and axis labels to make the plot informative and publication-ready.

* plt.show() Displays the plot.

```python
import pandas as ps
import seaborn as sns
import matplotlib.pyplot as plt

df = ps.read_csv("psy.csv")
df.head()
df.info()
```

The  code creates two scatter plots using the `seaborn` library:

**Study time vs. Final Grade:** Shows how student study time relates to their final grade

Each point on the plots represents a student, allowing us to see if there are visual trends between these variables and final grades.

```python
# Set the visual style
sns.set_theme(style="whitegrid")

# Simple smooth curve (no points)
plt.figure(figsize=(7, 4))
sns.regplot(
    data=df,
    x="Ambient Noise (dB)",
    y="Oxygen Saturation (%)",
    scatter=False,   # hide points
    lowess=True,     # use a smooth non-linear curve
    line_kws={"color": "royalblue", "lw": 3}
)
plt.title("Study Time vs Final Grade", fontsize=14)
plt.xlabel("Study Time (1 = <2h, 4 = >10h)")
plt.ylabel("Final Grade (G3)")
plt.show()
```

## Pearson Coeficient

in Python (with pandas or scipy), you can compute the Pearson correlation coefficient between two columns very easily.

$$
r_{XY} = \frac{\text{cov}(X, Y)}{\sigma_X \, \sigma_Y}
$$
R-squared (also called the coefficient of determination) tells you how much of the changes in
ð‘Œ
(for example, studentsâ€™ grades, stress levels, or satisfaction) can be explained by the predictors in your model.

In simple comparison we have :
$$R-squared = r^2$$



```python
r = df["Ambient Noise (dB)"].corr(df["Oxygen Saturation (%)"])
print("Pearson correlation coefficient:", r)
```

```python
rsquared = r**2
print("R-squared:", rsquared)
```

## Multi Linear Regression

Before going with MLR, we should only consider numeric attributes. It is very easy using pandas:


```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df.head()
```

We can now define ower MLR:

```python
import statsmodels.api as sm

# Prepare data
Y = numeric_df["Focus Duration (s)"]
X = numeric_df.drop(columns=["Focus Duration (s)"])
X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()
print(model.summary())
```

## Correlation Matrix

We can compute the correlation matrix to analyse and understand the quality of results

```python
corr = numeric_df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlation Matrix")
plt.show()
```

## Principal component analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.linear_model import LinearRegression

# Prepare data


# Apply PCA (1 component)
pca = PCA(n_components=1)
X_pca = pca.fit_transform(X)

# Fit linear regression
model = LinearRegression()
model.fit(X_pca, Y)

# Predict grades
y_pred = model.predict(X_pca)

# Sort points for smooth plotting
order = np.argsort(X_pca[:,0])
X_sorted = X_pca[order,0]
y_sorted = Y[order]
y_pred_sorted = y_pred[order]

# --- Plot ---
plt.figure(figsize=(9,6))
plt.scatter(X_pca, Y, color='gray', alpha=0.6, label='Actual data')
plt.plot(X_sorted, y_pred_sorted, color='blue', linewidth=2, label='Regression line')

plt.xlabel("Principal Component 1 (PCA summary of study habits, family, etc.)")
plt.ylabel("Final Grade (G3)")
plt.title("Linear Regression on First PCA Component (Student Performance)")
plt.legend()
plt.grid(True)
plt.show()
```
