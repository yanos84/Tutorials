```python
# prompt: create a random pandas dataframe with ID (incremental), Nitrogen (between 0 and 100), soil ph (between 13 and 0), Chlorophill (between 0 and 100), enzyme activity (between 0 and 2) and biomass (between 30 and 300)

import pandas as pd
import numpy as np

# Set the number of rows in the DataFrame
num_rows = 10

# Create a DataFrame with incremental IDs
df = pd.DataFrame({'ID': range(1, num_rows + 1)})

# Generate random values for other columns
df['Nitrogen'] = np.random.randint(0, 101, num_rows)
df['Soil pH'] = np.random.uniform(0, 13, num_rows)
df['Chlorophyll'] = np.random.randint(0, 101, num_rows)
df['Enzyme Activity'] = np.random.uniform(0, 2, num_rows)
df['Biomass'] = np.random.randint(30, 301, num_rows)

# Print the DataFrame
df
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression



# Define predictors (X) and target variable (y)
X = df[['Nitrogen', 'Soil pH', 'Chlorophyll', 'Enzyme Activity']]
y = df['Biomass']

# Fit regression model
model = LinearRegression()
model.fit(X, y)

print(f'Biomass ={model.coef_[0]:.2f}* Nitrogen +{model.coef_[1]:.2f}*Soil pH + {model.coef_[2]:.2f}*Chlorophill + {model.coef_[3]:.2f}*Enzyme activity')

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

```

```python
# Predict biomass for a new set of biochemical values
new_data = np.array([[45, 6.3, 40, 1.15]])  # Example values
predicted_biomass = model.predict(new_data)
print(f'Predicted Biomass:, {predicted_biomass[0]:.2f}')
```
